```python
def fillna_with_mode(df):
    for column in df.columns:
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)
    return df

data_filled = fillna_with_mode(data)
column_sums = data_filled.sum(numeric_only=None)
print(column_sums)
```

```python
# Count the missing values in each column
missing_values = data.isnull().sum()

# Display the count of missing values
print(missing_values)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Distribution of accidents by day of the week with a different color palette
plt.figure(figsize=(10, 6))
sns.countplot(data=data_filled, x='Day_of_week', order=data_filled['Day_of_week'].value_counts().index, palette='viridis')
plt.title('Distribution of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 2. Distribution of accidents by age band of the driver
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Age_band_of_driver', order=data['Age_band_of_driver'].value_counts().index)
plt.title('Distribution of Accidents by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 3. Correlation heatmap
# Ensure numeric columns
numeric_columns = data_filled.select_dtypes(include=['number']).columns
corr_matrix = data_filled[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")  # Example color map 'YlGnBu'
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt


cause_of_accident = data['Cause_of_accident'].value_counts().nlargest(10)

# Plotting
plt.figure(figsize=(10, 6))
colors = plt.cm.tab20(range(10))  # Choose a colormap and get 10 distinct colors
cause_of_accident.sort_values().plot(kind='barh', color=colors)
plt.title('Top 10 Causes of Accidents')
plt.xlabel('Accident Count')
plt.ylabel('Cause')
plt.gca().invert_yaxis()  # Invert y-axis to display the most common cause at the top
plt.show()
```

```python
import matplotlib.pyplot as plt
accident_severeness = data['Accident_severity'].value_counts()
plt.figure(figsize=(8, 6))
accident_severeness.plot(kind='bar', color='skyblue')  # Using a specific color for the bars
plt.xlabel("Seriousness of the Injury")
plt.ylabel("Count")
plt.title("Accident Severity")
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("task5.csv")

# Fill missing values with mode (most frequent value) for each column
data_filled = data.apply(lambda x: x.fillna(x.mode()[0]))

# Set up the figure size for all plots (increased size)
plt.figure(figsize=(18, 12))  # Increased figure size

# Plot 4: Distribution of accidents by type of vehicle
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot position 1
sns.countplot(data=data_filled, x='Type_of_vehicle', order=data_filled['Type_of_vehicle'].value_counts().index, palette='viridis')
plt.title('Distribution of Accidents by Type of Vehicle')
plt.xlabel('Type of Vehicle')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

# Plot 5: Distribution of accidents by weather conditions
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot position 2
sns.countplot(data=data_filled, x='Weather_conditions', order=data_filled['Weather_conditions'].value_counts().index, palette='viridis')
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

# Plot 6: Distribution of accidents by light conditions
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot position 3
sns.countplot(data=data_filled, x='Light_conditions', order=data_filled['Light_conditions'].value_counts().index, palette='viridis')
plt.title('Distribution of Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Show all plots
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("task5.csv")

# Fill missing values with mode (most frequent value) for each column
for column in data.columns:
    mode_value = data[column].mode()[0]
    data[column].fillna(mode_value, inplace=True)


# Additional plots

# 4. Distribution of accidents by type of vehicle
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Type_of_vehicle', order=data['Type_of_vehicle'].value_counts().index, palette='Set2')
plt.title('Distribution of Accidents by Type of Vehicle')
plt.xlabel('Type of Vehicle')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 5. Distribution of accidents by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Weather_conditions', order=data['Weather_conditions'].value_counts().index, palette='Set3')
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 6. Distribution of accidents by light conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Light_conditions', order=data['Light_conditions'].value_counts().index,palette='Pastel1')
plt.title('Distribution of Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()


```

```python
import matplotlib.pyplot as plt

# Count occurrences of each cause
cause_counts = data['Cause_of_accident'].value_counts()

plt.figure(figsize=(10, 8))  # Adjusted figsize for better proportions
colors = plt.cm.Paired(range(len(cause_counts)))  # Using a colormap for better color variation

plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('Proportion of Accidents by Cause', pad=20)  # Added padding to title for better spacing
plt.axis('equal')  # Ensures the pie chart is drawn as a circle

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data generation (replace with your actual data)
data = pd.DataFrame({
    'Time': ['08:30', '13:45', '17:20', '21:10', '09:15', '14:50', '18:30', '22:40', '07:00', '12:20'],
    'Cause_of_accident': ['Speeding', 'Distracted driving', 'Weather conditions', 'Speeding', 'Distracted driving',
                          'Weather conditions', 'Speeding', 'Distracted driving', 'Weather conditions', 'Speeding']
})

# Convert 'Time' to a categorical variable (e.g., time of day)
data['Time_of_day'] = pd.cut(pd.to_datetime(data['Time']).dt.hour, bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'])

plt.figure(figsize=(14, 7))
sns.countplot(data=data, x='Cause_of_accident', hue='Time_of_day', order=data['Cause_of_accident'].value_counts().index, palette='Set2')  # Using a different palette for better color differentiation
plt.title('Number of Accidents by Cause and Time of Day')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Time of Day')
plt.tight_layout()  # Ensures labels fit within the figure boundaries
plt.show()
```
